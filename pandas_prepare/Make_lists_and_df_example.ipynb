### Сначала делаю списки "синонимов"

#### Делаю два списка: 
    1) глаголов третьего лица наст ед числа наст времени
    2) существительных ед ч

##### Список глаголов

```python
# фильтрую по грам признакам и беру только уникальные слова
verbs3 = df1[df1.gram_features=="OrderedDict([('Mood', 'Ind'), ('Number', 'Sing'), ('Person', '3'), ('Tense', 'Pres'), ('VerbForm', 'Fin')])"][df1.pos == 'VERB'].word_form.unique()
```

```python
# понижаю регистр
verbs3 = [i.lower() for i in verbs3]
```

```python
# снова беру только уникальные слова (т к одни и те же глаголы были с большой и маленькой буквы)
verbs3 = list(set(verbs3))
```

##### Список существительных

```python
nouns_sing = df1[df1.pos=='NOUN'][df1.gram_features=="OrderedDict([('Number', 'Sing')])"].word_form.unique()
```

```python
nouns_sing = [i.lower() for i in nouns_sing]
nouns_sing = list(set(nouns_sing))
nouns_sing
```

### Делаю датафрейм, с которым сможет работать код бота
В датафрейме нужны следущие колонки:
 - текст для юзера с пропущенным словом;
 - правильный ответ;
 - название списка, из которого брать синонимы


```python
df_bot = df[['w_form', 't_for_user', 'pos', 'feats']] # беру нужные колонки из датафрейма
```

```python
df_bot = df_bot[(df_bot.pos=='NOUN')|(df_bot.pos=='VERB')] # фильтрую, чтобы остались существительные и глаголы
```

```python
# фильтрую, чтобы они остались только в нужной форме
df_bot = df_bot[(df_bot.feats=="OrderedDict([('Number', 'Sing')])")|(df_bot.feats=="OrderedDict([('Mood', 'Ind'), ('Number', 'Sing'), ('Person', '3'), ('Tense', 'Pres'), ('VerbForm', 'Fin')])")]
df_bot
```

```python
# понижаю регистр в колонке с правильными ответами
df_bot.w_form = df_bot.w_form.str.lower()
```

```python
df_bot
```

```python
# пишу функцию, которая будет возвращать название списка (списки я сделала и назвала выше)
def determine_gram_column(row):
    if row.pos == 'NOUN' and row.feats == "OrderedDict([('Number', 'Sing')])":
        return 'nouns_sing'
    elif row.pos == 'VERB' and row.feats == "OrderedDict([('Mood', 'Ind'), ('Number', 'Sing'), ('Person', '3'), ('Tense', 'Pres'), ('VerbForm', 'Fin')])":
        return 'verbs3'
     
```

```python
df_bot['list_syn'] = df_bot.apply(determine_gram_column, axis = 1) # применяю функцию
df_bot.drop(['pos','feats'], axis=1, inplace = True) # избавляюсь от уже ненужных колонок
```

```python
df_bot.reset_index( drop = True, inplace=True) # сбрасываем индекс
df_bot
```

```python
df_bot.to_csv('df_bot_1.csv', sep='$', encoding='utf-8', index=False) # записываем датафрейм в цсв
```
